```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import RandomOverSampler
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('thyroiddataset.csv')
df
```

```python
print(df.isnull().sum())
```

### Observations
* There are no missing values in the dataset 
* Since almost all the features are object types (categorical), we need to convert them to numerical values so machine learning models can work with them.

### One-Hot Encoding

```python
# Target column to binary
df['Recurred'] = df['Recurred'].map({'No': 0, 'Yes': 1})
```

```python
# Check original class balance

print("Original class balance:")
print(df['Recurred'].value_counts())
```

```python
# Encoding maps 

encoding_maps = {
    'Risk': {'Low': 0, 'Intermediate': 1, 'High': 2},
    'Response': {'Excellent': 0, 'Indeterminate': 1, 'Biochemical Incomplete': 2, 'Structural Incomplete': 3},
    'T': {'T1a': 0, 'T1b': 1, 'T2': 2, 'T3a': 3, 'T3b': 4, 'T4a': 5, 'T4b': 6},
    'N': {'N0': 0, 'N1a': 1, 'N1b': 2},
    'M': {'M0': 0, 'M1': 1},
    'Stage': {'I': 0, 'II': 1, 'III': 2, 'IVA': 3, 'IVB': 4},
    'Pathology': {'Papillary': 0, 'Micropapillary': 1, 'Follicular': 2, 'Hurthel cell': 3},
    'Focality': {'Uni-Focal': 0, 'Multi-Focal': 1},
    'Adenopathy': {'No': 0, 'Right': 1, 'Left': 2, 'Bilateral': 3, 'Extensive': 4, 'Posterior': 5},
    'Smoking': {'No': 0, 'Yes': 1},
    'Hx Smoking': {'No': 0, 'Yes': 1},
    'Hx Radiothreapy': {'No': 0, 'Yes': 1},
    'Thyroid Function': {
        'Euthyroid': 0,
        'Clinical Hyperthyroidism': 1,
        'Clinical Hypothyroidism': 2,
        'Subclinical Hyperthyroidism': 3,
        'Subclinical Hypothyroidism': 4
    },
    'Physical Examination': {
        'Multinodular goiter': 0,
        'Single nodular goiter-right': 1,
        'Single nodular goiter-left': 2,
        'Normal': 3,
        'Diffuse goiter': 4
    },
    'Gender': {'F': 0, 'M': 1}
}
```

```python
# Encode categorical features

for col, mapping in encoding_maps.items():
    df[col] = df[col].map(mapping)
```

```python
# Check for any NaNs after encoding and fill with 0 (safe fallback)

df = df.fillna(0)
```

```python
# Separate features and target

X = df.drop('Recurred', axis=1)
y = df['Recurred']
```

```python
# Check class balance before resampling

print("\nClass balance before resampling:")
print(y.value_counts())
```

```python
### Class Balance Before Resampling

plt.figure(figsize=(5,4))
sns.countplot(x='Recurred', data=df)
plt.title('Class Balance Before Resampling')
plt.xlabel('Recurrence (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Balance the classes

ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X, y)
```

```python
# Check class balance after resampling

print("\nClass balance after resampling:")
print(pd.Series(y_res).value_counts())
```

```python
### Class Balance after Resampling

plt.figure(figsize=(5,4))
sns.countplot(x=y_res)
plt.title('Class Balance After Resampling')
plt.xlabel('Recurrence (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()
```

### Observations 
* After resampling, the class distribution is balanced 
* The number of recurrence and non-recurrence cases is now equal. 
* This helps prevent model bias toward the majority class and improves prediction accuracy for both outcomes.

```python
# Train the model

rf = RandomForestClassifier(random_state=42)
rf.fit(X_res, y_res)
```

### Observations
* Created a Random Forest model and trained it on balanced data(resampled data).

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

### Observation 
*  To properly evaluate modelâ€™s performance (like with a confusion matrix, accuracy, etc.), we should split data into training and testing sets. 
* This helps to see how well model works on new, unseen data. 

### Feature Importances 

```python
importances = rf.feature_importances_
feat_names = X_res.columns

plt.figure(figsize=(8,5))
sns.barplot(x=importances, y=feat_names, orient='h')
plt.title('Feature Importances (Random Forest)')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = rf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Recurrence', 'Recurrence'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

### ROC Curve

```python
from sklearn.metrics import roc_curve, auc

y_score = rf.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_score)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
